```python
import skimage
astronaut = skimage.data.astronaut()
print(type(astronaut))
print(astronaut.shape)
skimage.io.imshow(astronaut)
skimage.io.show()
```

```python
from skimage import filters
# Aplicar el filtro gaussiano
# Utiliza 'channel_axis=-1' para indicar que el ltimo eje es el de los canales de color
astronaut_gaussian = filters.gaussian(astronaut, sigma=1, channel_axis=-1)

# Mostrar la imagen original y la imagen filtrada
skimage.io.imshow(astronaut)
skimage.io.show()

skimage.io.imshow(astronaut_gaussian)
skimage.io.show()
```
