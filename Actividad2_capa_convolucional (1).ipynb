```python
import cv2
from scipy.ndimage import convolve


image = cv2.imread('C:/Users/yjzabalam/Downloads/image2.jpg')
image = cv2.resize(image, (800, 600))  # Cambia el tama√±o a tus necesidades


```

```python
import numpy as np
from matplotlib import pyplot as plt


# Define el kernel
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

# Aplica la convoluci√≥n usando el kernel
convolved_image = cv2.filter2D(image, -1, kernel)

# Muestra las im√°genes original y convolucionada usando matplotlib
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Imagen Original')
plt.subplot(122), plt.imshow(cv2.cvtColor(convolved_image, cv2.COLOR_BGR2RGB)), plt.title('Imagen Convolucionada')
plt.show()
```
