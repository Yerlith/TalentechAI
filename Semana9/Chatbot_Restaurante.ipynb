```python
import openai
from datetime import datetime

# Configura tu clave de API de OpenAI
openai.api_key = 'tu_clave_de_api_aqui'

def get_openai_response(prompt):
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=prompt,
        max_tokens=150
    )
    return response.choices[0].text.strip()

def handle_reservation():
    print("Por favor proporciona la siguiente información para la reserva:")
    date = input("Fecha (dd/mm/aaaa): ")
    time = input("Hora (hh:mm): ")
    num_people = input("Número de personas: ")
    name = input("Nombre completo: ")
    print(f"Reserva confirmada para {num_people} persona(s) el {date} a las {time}. Nombre: {name}")

def handle_feedback():
    rating = input("Califica tu experiencia del 1 al 5: ")
    comment = input("Comentario adicional: ")
    print(f"Gracias por tus comentarios. Calificación: {rating}, Comentario: {comment}")

def chat():
    print("¡Hola! Bienvenido a [Nombre del Restaurante]. ¿En qué puedo ayudarte hoy?")
    
    while True:
        print("\nOpciones:")
        print("1. Hacer una reserva")
        print("2. Dejar una opinión")
        print("3. Preguntas frecuentes")
        print("4. Salir")
        
        choice = input("Selecciona una opción (1/2/3/4): ")
        
        if choice == '1':
            handle_reservation()
        elif choice == '2':
            handle_feedback()
        elif choice == '3':
            print("¿Qué información te gustaría saber?")
            print("1. Horarios de apertura")
            print("2. Menú y precios")
            print("3. Ubicación del restaurante")
            print("4. Política de cancelación")
            info_choice = input("Selecciona una opción (1/2/3/4): ")
            
            if info_choice == '1':
                prompt = "Proporcióname los horarios de apertura del restaurante."
                print(get_openai_response(prompt))
            elif info_choice == '2':
                prompt = "Proporcióname el menú y los precios del restaurante."
                print(get_openai_response(prompt))
            elif info_choice == '3':
                prompt = "Proporcióname la ubicación del restaurante."
                print(get_openai_response(prompt))
            elif info_choice == '4':
                prompt = "Proporcióname la política de cancelación del restaurante."
                print(get_openai_response(prompt))
            else:
                print("Opción no válida.")
        elif choice == '4':
            print("Gracias por tu visita. ¡Que tengas un buen día!")
            break
        else:
            print("Opción no válida. Por favor, selecciona una opción válida.")
            
if __name__ == "__main__":
    chat()
```
